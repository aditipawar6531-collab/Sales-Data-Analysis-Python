```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    "Product": ["Laptop", "Mobile", "Chair", "Tablet", "Mouse", "Laptop", "Mobile"],
    "Sales": [50000, 20000, 5000, 15000, 2000, 30000, 10000],
    "Region": ["West", "East", "South", "North", "West", "East", "South"]
}

df = pd.DataFrame(data)

print("Dataset:\n", df)

# 🔹 Total Sales
print("\nTotal Sales:", df["Sales"].sum())

# 🔹 Product-wise Sales
product_sales = df.groupby("Product")["Sales"].sum()
print("\nProduct-wise Sales:\n", product_sales)

# 🔹 Region-wise Sales
region_sales = df.groupby("Region")["Sales"].sum()
print("\nRegion-wise Sales:\n", region_sales)

# 🔹 Top 3 Products
top_products = product_sales.sort_values(ascending=False).head(3)
print("\nTop 3 Products:\n", top_products)

# 🔹 Highest Sales Region
top_region = region_sales.idxmax()
print("\nTop Region:", top_region)

# 📊 Graph 1: Product Sales
product_sales.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()

# 📊 Graph 2: Region Sales (Pie Chart)
region_sales.plot(kind="pie", autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```
